```python
r = requests.get('https://www.goodreads.com/book/show/1453138.Le_Lion')
soup = BeautifulSoup(r.content, 'html.parser')
```

```python
soup
```

```python
title = soup.find(attrs={"property": "og:title", })['content']
author = soup.find('a', class_ = "authorName").getText()
image = soup.find(attrs={"property": "og:image", })['content']
page_count = soup.find(attrs={"property": "books:page_count", })['content']
```
